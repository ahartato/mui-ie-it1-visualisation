## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Load dataset

```python
socmed_health_df = pd.read_csv("datasets/stud_socmed.csv")
```

```python
socmed_health_df.dtypes
```

## 3. Filter required columns

```python
selected_columns = [
    'Age',
    'Gender',
    'Academic_Level',
    'Country',
    'Avg_Daily_Usage_Hours',
    'Mental_Health_Score',
    'Sleep_Hours_Per_Night'
]

filtered_socmed_df = socmed_health_df[selected_columns]
print(filtered_socmed_df)
```

## 4. Create scatter plot

```python
plt.figure(figsize=(8, 6))

# Use reversed red colormap
scatter = plt.scatter(
    filtered_socmed_df['Avg_Daily_Usage_Hours'],
    filtered_socmed_df['Sleep_Hours_Per_Night'],
    c=filtered_socmed_df['Mental_Health_Score'],
    cmap='Reds_r',  # reversed Reds colormap
    alpha=0.7
)

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Mental Health Score')

# Add labels and title
plt.xlabel('Average Daily Usage (Hours)')
plt.ylabel('Sleep Hours Per Night')
plt.title('Usage vs Sleep (Darker Red = Lower Mental Health)')

plt.grid(True)
plt.show()
```

## 5. Export socmed data as csv

```python
# Export to CSV
filtered_socmed_df.to_csv('socmed_vs_mental.csv', index=False)
```
